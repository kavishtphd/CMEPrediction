```python
hist = 0
from PIL import Image 
import os
bins = np.linspace(0, 256, 65)
h = np.zeros(len(bins)-1)
for filename in os.listdir('FINALDATA/PNGHist/B'):
    #print(filename)
    gray_img = Image.open(os.path.join('FINALDATA/PNGHist/B', filename))
    h += np.histogram(np.array(gray_img).flatten(), bins)[0]
h[0] = 0
h /= h.sum()
```

```python
plt.bar(bins[:-1], h)
plt.show()
```

```python
bins = np.linspace(0, 256, 65)
h1 = np.zeros(len(bins)-1)
for filename in os.listdir('FINALDATA/PNGHist/M'):
    #print(filename)
    gray_img = Image.open(os.path.join('FINALDATA/PNGHist/M', filename))
    h1 += np.histogram(np.array(gray_img).flatten(), bins)[0]
h1[0] = 0
h1 /= h1.sum()

```

```python
plt.bar(bins[:-1]+2, h)
plt.bar(bins[:-1], h1)
plt.show()
```
