```python
import csv
import pandas as pd
import numpy as np
from datetime import datetime
sharpnumber =1
```

```python
statArray =  np.empty([0, 3])
```

```python
def getStats(sharpnum):
    #dfSharpRaw = pd.DataFrame()
    #print('SHARPDataOrig/SHARPData'+str(sharpnum)+'.csv')
    dfSharpRaw = pd.read_csv("SHARPDataOrig/SHARPData"+str(sharpnum)+".csv")
    dfSharpRaw.replace('MISSING', np.nan, inplace = True)
    #print(dfSharpRaw.head())
    time_i = pd.Series(dfSharpRaw.head(1)['T_OBS'])[0]
    #print(time_i)
    time_f = dfSharpRaw.tail(1)['T_OBS'][len(dfSharpRaw)-1]
    #print(time_f)
    datetime_i = datetime.strptime(time_i, '%Y.%m.%d_%H:%M:%S_TAI')
    datetime_f = datetime.strptime(time_f, '%Y.%m.%d_%H:%M:%S_TAI')
    datetime_d = (datetime_f-datetime_i).total_seconds()
    #print(datetime_d)
    mpot_max = max(dfSharpRaw['MEANPOT'].dropna())
    #mpot_mean = np.mean(dfSharpRaw['MEANPOT'].dropna())
    #print(mpot_mean)
    #print(mpot_mean)
    #print(mpot_max)
    return mpot_max, datetime_d
```

```python
while sharpnumber <= 7115:
    #print('heyyyy')
    try:
        print(sharpnumber)
        #print('tryyying'+'')
        mpot_m, datetime_diff = getStats(sharpnumber)
        #print('done with getstats')
        #print(datetime_diff)
    except FileNotFoundError:
        sharpnumber = sharpnumber + 1
        #print('suck it expetion')
        continue
    #print('out of try catch')
    statArray = np.append(statArray, [[sharpnumber, mpot_m, datetime_diff]], axis=0)
    sharpnumber = sharpnumber + 1
```

```python
dfSharpStats = pd.DataFrame(statArray, columns=['SHARP#', 'MAXMPOT', 'ElapsedTime'])
#dfSharpStats.set_index('SHARP#')
print(dfSharpStats)
```
