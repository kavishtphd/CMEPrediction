```python
while i<=7115:
    webdata = http.request('GET','http://jsoc.stanford.edu/cgi-bin/ajax/jsoc_info?ds=hmi.sharp_cea_720s['+str(i)+']&op=rs_list&key=MEANPOT,T_OBS')
    data = webdata.data
    dataJ = json.loads(data)
    try:
        l = int(len(dataJ["keywords"][0]["values"]))
    except KeyError:
        #print("now continuing")
        i=i+1
        continue

    with open('SHARPDataOrig/SHARPData'+str(i)+'.csv', 'w') as outFile:
        #print("now opening file")
        csvWriter = csv.writer(outFile, delimiter=',',
                            quotechar=' ', quoting=csv.QUOTE_MINIMAL)
        csvWriter.writerow(["SHARP#","MEANPOT","T_OBS"])
        x=0

        #print("now writing in the file")
        while x<l:
            csvWriter.writerow([str(i),dataJ["keywords"][0]["values"][x],dataJ["keywords"][1]["values"][x]])
            x=x+1
        #print("donw writing now closing")
        outFile.close()
    print("currently at: " + str(i))
    i=i+1
```
